```python
# # # Creates dialogue to ask directory
# # # Get the folder containing the image stack. 
root = tk.Tk()
root.attributes('-topmost', True)
root.withdraw() # Stops a second window opening
stack_file_name = filedialog.askopenfilename(title = 'Select Stack file')

print(stack_file_name)
```

```python
# Extract a single file name. 
im_data = czi.imread(stack_file_name)[0, 0, 0, 0, 0, 0, ..., 0]

print(im_data.shape)
```

**The image shape:**

```python
viewer = napari.Viewer()

viewer.add_image(im_data[0, ...])
viewer.add_image(im_data[1, ...])
```

```python
single_frame = im_data[0, 4, 5,...]
```

```python
def get_labels_for_z_pos(im_dat):
    
    blurred_im = skimage.filters.gaussian(im_dat, sigma =2)*100

    threshold = skimage.filters.threshold_otsu(blurred_im)

    # binarise image 
    thresh_im = np.array(blurred_im)

    thresh_im[thresh_im<threshold] = 1
    thresh_im[thresh_im>=threshold] = 0

    thresh_im = sci.ndimage.binary_fill_holes(thresh_im)


    labels = skimage.measure.label(thresh_im)
    
    return(labels)
    
# plt.imshow(thresh_im)
# plt.imshow(labels, alpha = 0.5)
# plt.colorbar()
```

```python
def filter_labels(labs):
    
    props = {'Label', 'Area'}

    label_props = pd.DataFrame(skimage.measure.regionprops_table(labs, properties= props))

    cutoff = 2000

    new_labels = np.zeros([labs.shape[0], labs.shape[1]], dtype = int)

    for i in range( label_props.shape[0]):
        mask_coords = np.where(labels == i)
        if len(mask_coords[0]) > cutoff: 
            new_labels[mask_coords[0], mask_coords[1]] = int(i)
            
    return(new_labels)
```

```python
frames_to_analyse = []

all_labels = []

for i in range(im_data.shape[2]):
    if np.mean(im_data[0, 0, i, ...]) > 8000: 
        single_frame = np.array(im_data[0, 0, i, ...])
        frames_to_analyse.append(i)
        
        labels = get_labels_for_z_pos(single_frame)
        
        filtered_labels = filter_labels(labels)
        
        all_labels.append(filtered_labels)
        
all_labels = np.array(all_labels)
```

```python
viewer = napari.Viewer()

viewer.add_image(im_data[0, 0, 4:, ...])
viewer.add_labels(all_labels)
```

```python
props = {'Label', 'Area'}
filtered_label_props = pd.DataFrame(skimage.measure.regionprops_table(filtered_labels, properties= props))
print(filtered_label_props)
```

```python

for j in range(6):# filtered_label_props.shape[0] ): 
    # Get image label ID. 
    label_val = filtered_label_props['Label'].iloc[j]
    
    # Find the pixel positions of the mask.
    pixels_loc = np.where(filtered_labels == label_val)
    
    square = [np.min(pixels_loc[0]), np.max(pixels_loc[0]), 
              np.min(pixels_loc[1]), np.max(pixels_loc[1])]
    
square_im = single_frame[square[0]:square[1], square[2]:square[3]]
cropped_labels = filtered_labels[square[0]:square[1], square[2]:square[3]]
    
plt.imshow(single_frame[square[0]:square[1], square[2]:square[3]])
plt.imshow(filtered_labels[square[0]:square[1], square[2]:square[3]], alpha = 0.5)

circ_label = get_labels_for_z_pos(square_im)
```
